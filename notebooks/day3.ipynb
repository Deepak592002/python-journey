```python
#palindrome
text=input("enter a str").lower()
strRev=str[::-1]
if(text==strRev):
    print("palindrome")
else:
    print("not palindrome")
```

```python
#string compress
text=input("enter a str :: ")
cnt=1
strComp=""
for i in range(1,len(text)):
    if(text[i-1]==text[i]):
        cnt+=1
    else:
        print(i-1)
        strComp+=text[i-1]+f"{cnt}"
        cnt=1
strComp+=text[-1]+f"{cnt}"
print(f"strComp :: {strComp}")
```
