```python
#string compress
text=input("enter a str :: ")
cnt=1
strComp=""
for i in range(1,len(text)):
    if(text[i-1]==text[i]):
        cnt+=1
    else:
        print(i-1)
        strComp+=text[i-1]+f"{cnt}"
        cnt=1
strComp+=text[-1]+f"{cnt}"
print(f"strComp :: {strComp}")
```

```python
#merge list
l1=list(map(int,list(input("enter list 1 ele :: "))))
l2=list(map(int,list(input("enter list 2 ele :: "))))
lenL1=len(l1)
lenL2=len(l2)
i=0
l=0
lMerge=[]
while(i<lenL1 and l<lenL2):
    if(l1[i]<=l2[l]):
        lMerge.append(l1[i])
        i+=1
    elif(l1[i]>=l2[l]):
        lMerge.append(l2[l])
        l+=1
while(i<lenL1):
    lMerge.append(l1[i])
    i+=1
while(l<lenL2):
    lMerge.append(l2[i])
    l+=1
print(lMerge)
```

```python
#remove duplicates
l=list(map(int,list(input())))
s=set()
l1=[]
for i in l:
    if i not in s:
        s.add(i)
        l1.append(i)

print(l1)
```

```python
#move zeros to end
l=list(map(int,list(input())))
c=0
for i in range(0, len(l)):
    if(l[i]==0):
        c+=1
for i in range(0,c):
    l.remove(0)
    l.append(0)
print(l)
```

```python
rows=int(input("enter rows"))
list=[]
for i in range(0, rows):
    ll=[]
    for l in range(0, rows):
        ll.append(int(input()))
    list.append(ll)
tl=[]

for i in range(0,rows):
    ll=[]
    for l in range(0, rows):
        ll.append(list[l][i])
    tl.append(ll)
for i in list:
    for l in i:
        print(l,end=" ")
    print()
    
for i in tl:
    for l in i:
        print(l, end=" ")
    print()
    
```
